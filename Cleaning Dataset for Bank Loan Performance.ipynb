Categorical Columns

```python
M = []
for i in df.columns :
  if df[i].dtype == 'object' :
    M.append(i)
print(M)
```

Label Encoding

```python
LE = LabelEncoder()
for i in M :
  df[i] = LE.fit_transform(df[i].astype(str))
```

train _test_Split

```python
x = df.drop('LOAN_STATUS',axis=1)
x
```

```python
y = df['LOAN_STATUS']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
rf = RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
# ---Train the Model---
P = ['ISSUE_DATE', 'LAST_CREDIT_PULL_DATE', 'LAST_PAYMENT_DATE', 'NEXT_PAYMENT_DATE']
x_train_processed = x_train.drop(columns = P)
rf.fit(x_train_processed,y_train)
```

```python
# ---Prediction Score---
P = ['ISSUE_DATE', 'LAST_CREDIT_PULL_DATE', 'LAST_PAYMENT_DATE', 'NEXT_PAYMENT_DATE']
x_test_processed = x_test.drop(columns = P)
y_pred = rf.predict(x_test_processed)
print(y_pred)
```

```python
y_pred = rf.predict(x_test_processed)
y_pred
```

```python
# ---Accuracy Score---
accuracy = accuracy_score(y_test,y_pred)
print('Accuracy :',accuracy)
```

```python
accuracy = accuracy_score(y_test,y_pred)
print('Accuracy :{:.2f}%'.format(accuracy*100))
```

```python
# ---Classification Report---
print(classification_report(y_test,y_pred))
```

```python
#---Confusion Matrix---
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
Feature_Importance = pd.DataFrame({'Feature' : x_train_processed.columns,'Importance' : rf.feature_importances_})
Feature_Importance
```
